## 4. Calculate Daily Gains/Losses

Using Pandas, add a new column named `["close_change"]` to the dataframe that calculates the difference between daily closing prices for FTX. This will result in positive (growth) and negative (loss) values.

Note:
- Be careful about how you calculate this. Each row should calculate that difference from the row above it, which must be the previous calendar day.
- Hint: This is not complicated and does not require a for-loop.  See the Pandas review notebook for reference.

```python
ftx_Clean = ftx_cleaned.applymap(lambda x: x.replace('"','')if isinstance(x,str) else x)
```

```python
ftx_Clean['timeOpen'] = pd.to_datetime(ftx_Clean.timeOpen)
ftx_Clean['timeClose'] = pd.to_datetime(ftx_Clean.timeClose)
ftx_Clean['timeHigh'] = pd.to_datetime(ftx_Clean.timeHigh)
ftx_Clean['timeLow'] = pd.to_datetime(ftx_Clean.timeLow)
ftx_Clean['timestamp'] = pd.to_datetime(ftx_Clean.timestamp)
```

```python
ftx_Clean.sort_values(by='timeOpen',inplace=True)
```

```python
ftx_Clean.head(5)
```

```python
ftx_Clean['open'] = pd.to_numeric(ftx_Clean.open)
ftx_Clean['close'] = pd.to_numeric(ftx_Clean.close)
ftx_Clean['high'] = pd.to_numeric(ftx_Clean.high)
ftx_Clean['low'] = pd.to_numeric(ftx_Clean.low)
ftx_Clean['volume'] = pd.to_numeric(ftx_Clean.volume)
ftx_Clean['marketCap'] = pd.to_numeric(ftx_Clean.marketCap)
ftx_Clean.dtypes
```

```python
ftx_Clean['close_change'] = ftx_Clean['close'].diff()
```

```python
ftx_Clean.head(5)
```

## 5. Identify Growth Days

Based on the daily change in closing prices calculated above, identify the top 10 days of positive growth for FTX.

Display your results in a table showing both the rate of change and their dates.

```python
ftx_Clean_SortedTopTen = ftx_Clean[ftx_Clean['close_change'] > 0].nlargest(10, 'close_change')
print("Top 10 Days of Positive Growth")
ftx_Clean_SortedTopTen[['timestamp','close_change']].head(10)
```

## 6. Calculate the Percentage Rate of Change

Next, create a new column named `["rate_of_change"]` based on `["close_change"]`.

Use the Pandas Review notebook for reference. Again, this is an easy operation in Pandas and does not require a for-loop.

```python
ftx_Clean['rate_of_change'] = ftx_Clean['close_change'].diff()
```

```python
ftx_Clean.head(5)
```

## 7. Identify Days of Greatest Percentage Change

Based on the rate change column calculated above, identify the top 5 days of greatest positive rate change, and the top 5 days of greatest negative rate change.

Display your results in a table showing both the rate of change and their dates.

```python
ftx_Top5_Positive = ftx_Clean[ftx_Clean['rate_of_change'] > 0].nlargest(5, 'rate_of_change')
ftx_Top5_Negative = ftx_Clean[ftx_Clean['rate_of_change'] < 0].nsmallest(5, 'rate_of_change')

print("Top 5 Days of Positive Rate of Change")
ftx_Top5_Positive[['timestamp','rate_of_change']].head(5)
```

```python
print("Top 5 Days of Negative Rate of Change")
ftx_Top5_Negative[['timestamp','rate_of_change']].head(5)
```

Create two separate plots of the data -- one for the daily closing price and one for the daily gains/losses. These can be histograms or line plots.

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 5))
plt.plot(ftx_Clean['timestamp'], ftx_Clean['close'], marker='o', color='b', label='Closing Price')
plt.title('Daily Closing Price')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(ftx_Clean['timestamp'], ftx_Clean['close_change'], marker='o', color='r', label='Gains/Losses')
plt.title('Daily Gains/Losses')
plt.xlabel('Date')
plt.ylabel('Daily Gains/Losses')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

Combine the two plots from above into a single plot, displaying both together. Your plot should have two Y-axis scales, should differentiate the two values by color, and should include a legend.

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.plot(ftx_Clean['timestamp'], ftx_Clean['close_change'], marker='o', color='r', label='Gains/Losses')
ax1.set_xlabel('Date')
ax1.set_ylabel('Daily Gains/Losses', color='r')
ax1.tick_params(axis='y', labelcolor='r')
ax1.grid(True)
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
ax2.plot(ftx_Clean['timestamp'], ftx_Clean['close'], marker='o', color='b', label='Closing Price')
ax2.set_ylabel('Closing Price', color='b')
ax2.tick_params(axis='y', labelcolor='b')
ax2.legend(loc='upper right')

plt.title('Combined Plot: Daily Gains/Losses and Closing Price')

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```
